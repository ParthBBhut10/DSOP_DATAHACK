```python
def clean_text(text):
  for char in text:
    if char in string.punctuation:
      text = text.replace(char,'')
  return text
```

```python
video['title'] = video['title'].apply(lambda x:clean_text(x))
#video['description'] = video['description'].apply(lambda x:clean_text(x))
video['tags'] = video['tags'].apply(lambda x:clean_text(x))
video['Comments'] = video['Comments'].apply(lambda x:clean_text(x))
video
```

```python
from collections import Counter

video['title']
# Define the number of top words to show
top_n = 5

# Create a Counter object to count the occurrence of each word
word_counts = Counter()

for sentence in video['title']:
    # Split the sentence into words
    words = sentence.split()
    # Update the word_counts with the occurrence of each word in the sentence
    word_counts.update(words)

# Get the top n most common words and their frequencies
top_words = word_counts.most_common(top_n)

print(f"Top {top_n} most frequent words in the dataset:")
for word, frequency in top_words:
    print(f"'{word}': {frequency} occurrences")




```
