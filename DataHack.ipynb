Importing the Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
from mpl_toolkits.mplot3d import Axes3D
```

Loading the Dataset

```python
data=pd.read_csv('/content/dataset_datahack_videos.csv')
data
```

Spliting the upload_time to two different columns  date and time 

```python
data['date']=""
data['time']=""
for i in range(len(data['upload_time'])):
   data['date'][i] = data['upload_time'][i].split('T')[0]
   data['time'][i] = data['upload_time'][i].split('T')[1].split('Z')[0]
data
```

Converting time column from H M S format  to minutes

```python
from numpy import e
for i in range(len(data)):
  a=data.duration[i]
  a=a[2:]
  try:
    H=a.index("H")
  except:
    H=False
  try:
    M=a.index("M")
  except:
    M=False
  try:
    S=a.index("S")
  except:
    S=False
  if H and M and S:
     h=int(a[:H])
     m=int(a[H+1:M])
     s=int(a[M+1:-1])
     t=h*60+m+s/60
  elif H and M:
     h=int(a[:H])
     m=int(a[H+1:M])
     t=h*60+m
  elif H and S:
     h=int(a[:H])
     s=int(a[H+1:-1])
     t=h*60+s/60
  elif M and S:
     m=int(a[:M])
     s=int(a[M+1:-1])
     t=m+s/60
  elif H:
     h=int(a[:H])
     t=h*60
  elif M:
     m=int(a[:M])
     t=m
  elif S:
     s=int(a[:-1])
     t=s


    
  data.duration[i]=t
```

Plotting the Graph between the Channels and Views 

```python
df2 = data.groupby('channel_title').sum().sort_values('view_count',ascending = False)['view_count'].head(10)
plt.figure(figsize = (16,8))
df2.plot.bar().bar_label(df2.plot.bar().containers[0])
plt.xticks(rotation = 75)
plt.xlabel('Channel Title')
plt.ylabel('Views (in Crores)')
plt.ylim(0,150000000)
plt.title('Channels with most Views')
plt.show()
```

Plotting the Graph between likes and Channel

```python
df3 = data.groupby('channel_title').sum().sort_values('likes',ascending = False)['likes'].head(10)
plt.figure(figsize = (16,8))
df3.plot.bar().bar_label(df3.plot.bar().containers[0])
plt.ylim(0,9000000)
plt.xticks(rotation = 75)
plt.ylabel('Number of Likes (in  Millions)')
plt.xlabel('Channel Title')
plt.title('Channels with most Likes')
plt.show()
```

Plotting the Graph between Number of Comments and Channel

```python
df4 = data.groupby('channel_title').sum().sort_values('comments',ascending = False)['comments'].head(10)
plt.figure(figsize = (16,8))
df3.plot.bar().bar_label(df4.plot.bar().containers[0])
plt.ylim(0,9000000)
plt.xticks(rotation = 75)
plt.ylabel('Number of Comments (in  Millions)')
plt.xlabel('Channel Title')
plt.title('Channels with most Comments')
plt.show()
```

 Here with the above three Graphs we can observe that the Number of Views,Likes and Comments are more on the channel of R.G Bucket List

Assigning the group to the particular time period
